##Load data from ADLS (Azure DataLake Storage)
Add your credentials from your Storage Account (ADLS)

```
STORAGE_ACCOUNT_NAME = '<STORAGE_ACCOUNT_NAME>'
CONTAINER_INPUT = '<YOUR_CONTAINER>'
CONTAINER_OUTPUT = '<YOUR_CONTAINER>'
KEY = '<PRIMARY_KEY>'
```

##Mount ADLS
We will mount two folders (input and output files)

```
dbutils.fs.mount(
  source = "wasbs://{}@{}.blob.core.windows.net".format(CONTAINER_INPUT, STORAGE_ACCOUNT_NAME),
  mount_point = "/mnt/{}".format(CONTAINER_INPUT),
  extra_configs = {"fs.azure.account.key.{}.blob.core.windows.net".format(STORAGE_ACCOUNT_NAME):"{}".format(KEY)})
```

```
dbutils.fs.mount(
  source = "wasbs://{}@{}.blob.core.windows.net".format(CONTAINER_OUTPUT, STORAGE_ACCOUNT_NAME),
  mount_point = "/mnt/{}".format(CONTAINER_OUTPUT),
  extra_configs = {"fs.azure.account.key.{}.blob.core.windows.net".format(STORAGE_ACCOUNT_NAME):"{}".format(KEY)})
```

##Read from ADLS to Spark
Get data from ADLS and transform to Spark DataFrame

```
FILENAME = "/mnt/{}/UsedCars.csv".format(CONTAINER)

UsedCars = (spark
  .read
  .option("header", True)
  .option("header", True) \
  .option("sep", ',') \
  .csv(FILENAME)
)
```

```
display(UsedCars)
```
