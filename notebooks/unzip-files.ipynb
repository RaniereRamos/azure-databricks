## How to unzip files from ADLS (Azure DataLake Storage)
We cannot unzip files directly from the ADLS. First, you need to copy locally (to Databricks filestore) to be able to unzip. We can use /tmp directory to do this task.

In this example I assume you already have mounted ADLS to Databricks (if you don't know how to do it, please see this [link](https://github.com/lfbraz/azure-databricks/blob/master/notebooks/read-from-adls.ipynb)).

## List mounted ADLS
Get the path of your zip file

```
dbutils.fs.ls('/mnt/YOUR_CONTAINER/')
```

## Copy your zip file (from ADLS to local databricks storage)

```
dbutils.fs.cp('dbfs:/mnt/YOUR_CONTAINER/YOUR_ZIP_FILE.zip', 'file:/tmp/YOUR_ZIP_FILE.zip', True)  
```

##List /tmp folder to check if you file was correctly copied

```
dbutils.fs.ls('file:/tmp/')
```

##Unzip the file
It's not necessary to put .zip extension for the filename. <br/>
-d option allow you to extract files to an specific directory.  <br/>
-o option allow you to replace existing files.

```
%sh
unzip -o /tmp/YOUR_FILE -d /tmp
```

##List /tmp folder to check if you file was correctly extracted

```
dbutils.fs.ls('file:/tmp/')
```

##Now you can copy extracted file to ADLS

```
dbutils.fs.cp('file:/tmp/YOUR_EXTRACTED_FILE.csv', 'dbfs:/mnt/YOUR_CONTAINER/YOUR_EXTRACTED_FILE.csv')  
```
